```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Variáveis categóricas
cols_categoricas = ['pclass', 'embarked_c', 'embarked_q', 'embarked_s', 'age_group', 
                    'is_male', 'cabin_group', 'title', 'fare_group', 'is_alone', 
                    'sex_pclass_interaction']

for col in cols_categoricas:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, hue='survived', data=df, palette='viridis')
    plt.title(f'Sobrevivência por {col}')
    plt.show()
```

## Conclusão
* `pclass`: Pessoas na primeira classe tinham mais chances de sobreviver.
* `embarked`: O porto C foi o único com maior chances de sobrevivência. 
* `age_group`: Claramente crianças foram retiradas primeiro.
* `is_male`: Mulheres também foram retiradas primeiro.
* `cabin_group`: Pessoas com os grupos C, E, D, B e F de cabine foram mais salvas do que não.
* `title`: Títulos femininos foram mais salvos, reforçando a política de mulheres e crianças primeiro.
* `fare_group`: Quanto mais alta a taxa mais sobreviveram.
* `is_alone`: Quem esatava sozinho morreu mais do que quem não estava.
* `sex_pclass_interaction`: Foram salvos pela ordem de mulheres da 1, 2 e 3 classe, em seguida de homens da 1, 3 e 2 classe. Não tendo um padrão tão claro em relação aos homens. 

```python
# Variáveis numéricas
cols_numericas = ['age', 'sibsp', 'parch', 'fare', 'family_size', 'ticket_length', 'ticket_qtd']

for col in cols_numericas:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data=df[df['survived'] == 0][col], label='Não Sobreviveu', shade=True)
    sns.kdeplot(data=df[df['survived'] == 1][col], label='Sobreviveu', shade=True)
    plt.title(f'Distribuição de {col} por sobrevivência')
    plt.legend()
    plt.show()
```
